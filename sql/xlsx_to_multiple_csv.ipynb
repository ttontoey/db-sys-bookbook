```python
from google.colab import files
import pandas as pd
import os
import shutil

uploaded = files.upload()

file_name = next(iter(uploaded))
excel_data = pd.read_excel(file_name, sheet_name=None, engine='openpyxl')

output_dir = '/content/csv_files'
os.makedirs(output_dir, exist_ok=True)  # Create directory if it doesn't exist

for sheet_name, df in excel_data.items():
    csv_file_path = os.path.join(output_dir, f'{sheet_name}.csv')
    df.to_csv(csv_file_path, index=False, header=False)

zip_file_path = '/content/csv_files.zip'
shutil.make_archive(zip_file_path.replace('.zip', ''), 'zip', output_dir)

files.download(zip_file_path)
```
